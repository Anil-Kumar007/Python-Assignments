```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print("Row-wise sum:", row_sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame?
import pandas as pd

df = pd.DataFrame({'A': [10, 20, 30], 'B': [15, 25, 35]})
mean_value = df['A'].mean()
print("Mean of column A:", mean_value)
```

```python
# 3. Create a scatter plot using Matplotlib?
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [5, 6, 7, 8]
})

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 5. Generate a bar plot using Plotly?
import plotly.express as px
import pandas as pd

df = pd.DataFrame({'Fruits': ['Apple', 'Banana', 'Cherry'], 'Count': [10, 15, 7]})
fig = px.bar(df, x='Fruits', y='Count', title="Fruit Count")
demonstrate = fig.show()
demonstrate
```

```python
# 6. Create a DataFrame and add a new column based on an existing column?
import pandas as pd

df = pd.DataFrame({'Value': [10, 20, 30]})
df['Double'] = df['Value'] * 2
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays?
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print("Element-wise multiplication:", result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib?
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 7]

plt.plot(x, y1, label='Squared')
plt.plot(x, y2, label='Other')
plt.legend()
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?
import pandas as pd

df = pd.DataFrame({'Score': [45, 67, 89, 32, 76]})
filtered_df = df[df['Score'] > 60]
print(filtered_df)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution?
import seaborn as sns
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
sns.histplot(data, bins=5, kde=True)
plt.title("Distribution Histogram")
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy?
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
product = np.dot(a, b)
print("Matrix multiplication result:\n", product)
```

```python
# 13. Create a 3D scatter plot using Plotly?
import plotly.express as px
import pandas as pd
import plotly.io as pio

# Set the renderer depending on your environment
pio.renderers.default = 'browser'  # use 'notebook' if using Jupyter

# Sample DataFrame
df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 15, 13, 17],
    'z': [5, 3, 6, 9]
})

# Create 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z', title="3D Scatter Plot")
fig.show()
```
